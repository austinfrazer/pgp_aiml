### 8. Visualize how popularity of genres has changed over the years - 10 marks

Note that you need to use the **percent of number of releases in a year** as a parameter of popularity of a genre

Hint 1: You need to reach to a data frame where the release year is the index and the genre is the column names (one cell shows the number of release in a year in one genre) or vice versa. (Drop unnecessary column if there are any)

Hint 2: Find the total number of movies release in a year(use item dataset to get count of movies released in a particular year, store that value in a new column as 'total'). Now divide the value of each genre in that year by total to get percentage number of release in a particular year.
`(df.div(df['total'], axis= 0) * 100)`

Once that is achieved, you can either use univariate plots or can use the heatmap to visualise all the changes over the years 
in one go. 

Hint 3: Use groupby on the relevant column and use sum() on the same to find out the number of releases in a year/genre.

```python
grouped_data = movies_df.iloc[:, np.r_[0, 3:20, 21]].groupby('release year')
total_movies_by_year = pd.DataFrame(grouped_data['movie id'].count()).rename(columns = ({'movie id': 'total_movies'}))
```

```python
grouped_data = movies_df.iloc[:, np.r_[3:20, 21]].groupby('release year')
genre_movies_by_year = grouped_data.sum()
```

```python
total_and_genre_by_year = pd.concat([genre_movies_by_year, total_movies_by_year], axis = 'columns')
```

```python
# Using the hint to get the percentage totals
genre_pct_by_year = total_and_genre_by_year.div(total_and_genre_by_year['total_movies'], axis= 0) * 100
# Dropping the unknown genre column since it is always 0% and the total column since it is always 100%
genre_pct_by_year = genre_pct_by_year.iloc[:, 1:17]
# This will put the most recent year first which will make more sense in the heatmap.
genre_pct_by_year = genre_pct_by_year[::-1]
```

```python
# For some of the graph types that I want to use, I need to rearrange the data into molten format (one value per row.)
molten_genre_pct_by_year = genre_pct_by_year.reset_index().melt(id_vars = ['release year']).rename(columns = ({'variable':'genre', 'value' : 'percent'}))
```

```python
molten_genre_pct_by_year.head(6).style.hide_index()
```

```python
print("A quick but messy look at all different genres of movie by year using lineplot().")
sns.lineplot(data = molten_genre_pct_by_year, x = 'release year', y = 'percent', hue = "genre", style = "genre");
```

```python
print("Taking a look at all genres of movies using the facet grid.")
g = sns.FacetGrid(data = molten_genre_pct_by_year, col="genre", col_wrap=4, height=4, sharex = False)
g = g.map(plt.plot, "release year", "percent" , marker="")
```

### 9. Find the top 25 movies in terms of average ratings for movies that have been rated more than 100 times - 10 marks

Hints : 

1. Find the count of ratings and average ratings for every movie.
2. Slice the movies which have ratings more than 100.
3. Sort values according to average rating such that movie which highest rating is on top.
4. Select top 25 movies.
5. You will have to use the .merge() function to get the movie titles.

Note: This question will need you to research about groupby and apply your findings. You can find more on groupby on https://realpython.com/pandas-groupby/.

```python
#    Find the count of ratings and average ratings for every movie.
#    Find the count of ratings
grouped_data = ratings_df.iloc[:, 1:3].groupby('movie id')
total_ratings_by_movie = grouped_data.count().rename(columns = ({'rating' : 'total ratings'}))
total_ratings_by_movie.head()
```

```python
#    Find the average ratings for every movie
average_ratings_by_movie = grouped_data.mean().rename(columns = ({'rating' : 'average ratings'}))
average_ratings_by_movie.head()
```

```python
#    Slice the movies which have ratings more than 100.
# Combining two grouped by dataframes.
ratings_by_movie = pd.concat([average_ratings_by_movie, total_ratings_by_movie], axis = 'columns')
# Getting the subset for movies with more than 100 ratings.
greater_than_100 = ratings_by_movie['total ratings'] > 100
# Subsetting.
top_100_movies = ratings_by_movie[greater_than_100]
top_100_movies.head()
```

```python
#    Sort values according to average rating such that movie which highest rating is on top.
sorted_top_100_movies = top_100_movies.sort_values(['average ratings'], ascending = False)
sorted_top_100_movies.head()
```

```python
#    Select top 25 movies.
sorted_top_25_movies = sorted_top_100_movies.iloc[0:25, :]
sorted_top_25_movies
```

```python
#    You will have to use the .merge() function to get the movie titles.
columns_of_interest = ['movie id', 'movie title']
key_variable = 'movie id'
sorted_top_25_movies_with_title = pd.merge(sorted_top_25_movies, movies_df[columns_of_interest], on = key_variable)

cols = sorted_top_25_movies_with_title.columns.tolist()
cols = cols[-1:] + cols[:-1]
sorted_top_25_movies_with_title = sorted_top_25_movies_with_title[cols]
print("I am leaving the movie id in order to be able to check the join.")
print("I am leaving the total ratings in order to affirm that only movies with > 100 ratings were chosen.\n\n")
sorted_top_25_movies_with_title.style.hide_index()
```

### 10. Check for the validity of the below statements with respect to the data provided - 10 marks

* Men watch more drama than women
* Women watch more Sci-Fi than men
* Men watch more Romance than women

**compare the percentages**

**Please pay attention to what should be the denominator while calculating percentages**

1. Merge all the datasets

2. There is no need to conduct statistical tests around this. Just **compare the percentages** and comment on the validity of the above statements.

3. you might want ot use the .sum(), .div() function here.

4. Use number of ratings to validate the numbers. For example, if out of 4000 ratings received by women, 3000 are for drama, we will assume that 75% of the women watch drama.

```python
#  Merge all the datasets, pt 1
key_variable = 'movie id'
r_m = pd.merge(ratings_df, movies_df, on = key_variable)
r_m.head(3).style.hide_index()
```

```python
#  Merge all the datasets, pt 2
key_variable = 'user id'
ratings_movies_user_df_merged = pd.merge(r_m, user_df, on = key_variable)
ratings_movies_user_df_merged.head(3).style.hide_index()
```

```python
print("Assertion 1:  Men watch more drama than women.")
genre_of_choice = 'Drama'
num_movies_women_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'F'])
num_movies_men_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'M'])
num_genre_women_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'F')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
num_genre_men_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'M')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
men_prop_genre = num_genre_men_watched / num_movies_men_watched
women_prop_genre = num_genre_women_watched / num_movies_women_watched

if men_prop_genre > women_prop_genre:
    print ("True, men watch more {} than women.".format(genre_of_choice))
else:
    print ("False, men do not watch more {} than women.".format(genre_of_choice))
print("Men watch {0}% {1}.  Women watch {2}% {1}.".format(str(round(men_prop_genre * 100, 2)),
                                                            genre_of_choice,
                                                            str(round(women_prop_genre * 100, 2))))
```

```python
print("Assertion 2:  Women watch more Sci-Fi than men.")
genre_of_choice = 'Sci-Fi'
num_movies_women_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'F'])
num_movies_men_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'M'])
num_genre_women_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'F')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
num_genre_men_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'M')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
men_prop_genre = num_genre_men_watched / num_movies_men_watched
women_prop_genre = num_genre_women_watched / num_movies_women_watched

if women_prop_genre > men_prop_genre:
    print ("True, women watch more {} than women.".format(genre_of_choice))
else:
    print ("False, women do not watch more {} than men.".format(genre_of_choice))
print("Men watch {0}% {1}.  Women watch {2}% {1}.".format(str(round(men_prop_genre * 100, 2)),
                                                            genre_of_choice,
                                                            str(round(women_prop_genre * 100, 2))))
```

```python
print("Assertion 3:  Men watch more Romance than women.")

genre_of_choice = 'Romance'
num_movies_women_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'F'])
num_movies_men_watched = len(ratings_movies_user_df_merged[ratings_movies_user_df_merged['gender'] == 'M'])
num_genre_women_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'F')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
num_genre_men_watched = len(ratings_movies_user_df_merged[(ratings_movies_user_df_merged['gender'] == 'M')
                                                             & (ratings_movies_user_df_merged[genre_of_choice] == 1)])
men_prop_genre = num_genre_men_watched / num_movies_men_watched
women_prop_genre = num_genre_women_watched / num_movies_women_watched

if men_prop_genre > women_prop_genre:
    print ("True, men watch more {} than women.".format(genre_of_choice))
else:
    print ("False, men do not watch more {} than women.".format(genre_of_choice))
print("Men watch {0}% {1}.  Women watch {2}% {1}.".format(str(round(men_prop_genre * 100, 2)),
                                                            genre_of_choice,
                                                            str(round(women_prop_genre * 100, 2))))
```
